```python
import json
import matplotlib.pyplot as plt
```

```python
DATA1_NAME = "8m_9m"
DATA2_NAME = "10m_11m_to_8m_9m"
```

```python
#### Draw graph ####
def plot_data(data1, data2, title_text):
    plt.title("Easy Scenarios")
    plt.plot(data1, label = DATA1_NAME)
    plt.plot(data2, label = DATA2_NAME)
    plt.xlabel("Step(Every 10000 timestamps)")
    plt.ylabel(title_text)
    plt.legend()
    figure_name = DATA1_NAME + "_" + DATA2_NAME + "_" + title_text + ".png"
    print(figure_name)
    plt.savefig(figure_name, dpi = 1000)
    plt.show()
```

```python
train_result_file_1 = open('8m_9m.json')
train_result_file_2 = open('10m_11.json')
data1 = json.load(train_result_file_1)
data2 = json.load(train_result_file_2)
```

```python
# battle_won_mean
plot_data(data1["battle_won_mean"], data2["battle_won_mean"], "battle_won_mean")
```

```python
# test_battle_won_mean:
# print(len(data2["test_battle_won_mean"]))
plot_data(data1["test_battle_won_mean"][:120], data2["test_battle_won_mean"][:120], "test_battle_won_mean")
```

```python
# dead_allies_mean
plot_data(data1["dead_allies_mean"], data2["dead_allies_mean"], "dead_allies_mean")
```

```python
# dead_enemies_mean
plot_data(data1["dead_enemies_mean"], data2["dead_enemies_mean"], "dead_enemies_mean")
```

```python
# loss
plot_data(data1["loss"], data2["loss"], "loss")
```

```python
# dead_allies_mean
plot_data(data1["dead_allies_mean"], data2["dead_allies_mean"], "dead_allies_mean")
```
